```python
from src.utils.folders_utils import get_repo_folder
import pandas as pd

repo_folder = get_repo_folder()
print(f"Repository folder: {repo_folder}")

hqa_path = repo_folder / "src/data/DB_questions/History/HQA-data/hqa_train_df.csv"

# show head of each train_data.csv file
hqa_df = pd.read_csv(hqa_path)
# shuffle the dataframe
hqa_df = hqa_df.sample(frac=1).reset_index(drop=True)
hqa_df.head(20)
```

```python
def format_text(text, n=100):
    # if its too long than n characters, split it into lines of n characters
    if len(text) > n:
        return '\n'.join(text[i:i+n] for i in range(0, len(text), n))
    return text

for index in range(5):
    # print for the first row the context, question, and answer. 
    print(f"Context: {format_text(hqa_df['Context'][index])}")
    print(f"Question: {format_text(hqa_df['Question'][index])}")
    print(f"Answer: {format_text(hqa_df['Answer'][index])}")
    print("-" * 50)
```

we can see that the context is very long, we won't use it, but only the question answer pairs
